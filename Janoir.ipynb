## Electrostatics

Force between swimmer and colloid,

$F = \dfrac{\mathrm{Z} \mathrm{R^*}}{\lambda_D} \mathrm{e}^{-d / \lambda_D}$

Reduced mass,

$\mathrm{R^*} = \dfrac{R R_c}{R + R_c}$

$d$ is the distance between the colloid's *surface* and the swimmer's *surface*.

For now keep electrostatic energy thing $Z \mathrm{R^*} = 10^{-4} \mathrm{\, kg \, {µm}^2 s^{-2}}$ constant. (This is units of energy but using µm.)

Typically $\lambda_D = 0.01 \mathrm{\, µm}$ but depends on salt.

```python
R_reduced = (R * R_c) / (R + R_c)
ZR_reduced = 1e-16
Z = ZR_reduced / R_reduced
debye_length = 0.01

print('Based on electrostatic energy constant, Z = {:.2g}'.format(Z))
```

## Hydrodynamics

Equations taken from Spagnolie and Lauga (2012) J. Fluid Mech. 700, 105

For a stokeslet dipole field.

Taking the colloid surface as a plane.

Prefactor in all hydrodynamic terms, 

$H = (3 / 16) \alpha R^2 v_p$

Angular rotation of swimmer's direction,

$\dot{\theta} = -2 H \dfrac{\sin{2 \theta}}{h^3}$

Perpendicular and parallel velocities,

$v_\perp = -2 H \dfrac{1 - 3 \sin^2{\theta}}{h^2}$

$v_\parallel = H \dfrac{\sin{2 \theta}}{h^2}$

$h$ is the distance between the colloid's *surface* and the swimmer's *centre*

Alpha controls the strength of the interaction, Aidan says for now set $\alpha = 1$.

```python
alpha = 1.0
H = (3.0 / 16.0) * alpha * R ** 2 * v_p
```

# Diffusion

From Wikipedia page on Einstein relation, in Stokes-Einstein section for low Reynolds number,

$D = \dfrac{k_B T}{6 \pi \eta R}$

$D_r = \dfrac{k_B T}{8 \pi \eta R^3}$

Boltzmann's constant has units of $\mathrm{m^2 kg s^{-2} K^{-1}} = 10^{12} \mathrm{{µm}^2 kg s^{-2} K^{-1}}$

```python
import scipy.constants
k_b = 1e12 * scipy.constants.k
T = 300.0
D = (k_b * T) / (6.0 * np.pi * eta * R)
D_r = (k_b * T) / (8.0 * np.pi * eta * R ** 3)

print('At T = {:.3g} K:'.format(T))
print('Translational diffusion constant D = {:.2g} microns^2/s'.format(D))
print('Rotational diffusion constant D_r = {:.2g} radians/s'.format(D_r))
```

In terms of implementation, for diffusion, each quantity is displaced in one time-step by a step with magnitude picked from a normal distribution with standard deviation $\sqrt{2 D dt}$.

For translation, the quantity is position, and $D$ is the translational diffusion constant $D$.

For rotation, the quantity is orientation, and $D$ is the rotational diffusion constant $D_r$.

In $n$ dimensions, have $n$ translational degrees of freedom. Displacement along these axes is trivial.

In $n$ dimensions, have $n (n - 1) / 2$ rotational degrees of freedom, from http://en.wikipedia.org/wiki/Degrees_of_freedom_(mechanics).

Displacement along these axes involves rotation about independent axes by the above amount.

This is more complicated in higher dimensions. But in 2 dimensions, have only 1 rotational degree of freedom. Can pick 1 angle and rotate by this angle.

Done using a random rotation matrix. In 2 dimensions this looks like:

\begin{bmatrix}
    \cos{\theta} & -\sin{\theta}  \\
    \sin{\theta} & \cos{\theta}
\end{bmatrix}

# 2D approximation

Because swimmers are confined to between the colloids and the wall, they move in a quasi-2D world. Can use this to simplify simulations without simulating the wall explicitly.

They see colloids as circular obstacles with some effective radius, $R_{c, eff} < R_c$.

From geometry, $R_{c, eff} = R_c \cos{\phi}$, where $\phi$ is the angle from the lattice axis on which the centre of the swimmer lies.

Assuming the colloid and swimmer lie adjacent to the wall, the distance from colloid centre to swimmer centre is $R_c + R$. The vertical distance between the swimmer centre and the horizontal line on which the colloid centre lies must be $R_c - R$. Therefore , $\sin{\phi} = \dfrac{R_c - R}{R_c + R}$.

```python
phi = np.arcsin((R_c - R) / (R_c + R))
R_c_eff = R_c * np.cos(phi)
print('Swimmer lies at lattice axis off-angle phi = {:.3g} degrees'.format((180.0 / np.pi) * phi))
print('So effective colloid radius seen by swimmer is R_c_eff = {:.3g} microns'.format(R_c_eff))
```

# Hexagonal lattice algorithm

Aidan's experiments involve fixed colloids on a hexagonal lattice.

From http://stackoverflow.com/q/14280831, algorithm for generating 2 dimensional hexagonal lattice. Converted into Python, for a lattice of order n:

```python
def hex_lattice(n):
    rs = []
    for i in range(n):
        y = np.sqrt(3) * i / 2.0
        for j in range(2 * n - 1 - i):
            x = (-2 * n + i + 2) / 2.0 + j
            rs.append([x, y])
            if y != 0.0:
                rs.append([x, -y])
    return rs
```

# Leaving time measurement

To measure how stuck the swimmers are to the colloids, carry out simulations using one swimmer and one colloid.

Initialise the swimmer so that the starting separation between swimmer and colloid surfaces, $d = d_0 = 3 \mathrm{\, µm}$.

Initialise the swimmer orientation on a tangential trajectory to graze the colloid. 

If and when $d = d_a = 1 \mathrm{\, µm}$, consider the swimmer as 'attached' at time $t_a$.

When $d > d_d = 2 \mathrm{\, µm}$, at time $t_d$, consider the active colloid as 'detached'.

The leaving time, $\tau = t_d - t_a$.

If the simulation time $t > 50 \mathrm{\, s}$, take $\tau = \infty$. 

The reason for $d_a < d_d$ is to prevent very small $\tau$ if the particle jumps into and out of the 'attachment region' quickly via diffusion.

Aidan suggested a better way to calculate the hopping rate than taking the median (which is unbiased but limited), which is unbiased in regards to there being an upper limit to how long I wait to see if the particle leaves.

Have $N$ simulation runs, and therefore $N$ potential hops.

Have $m$ runs where the particle doesn't leave within the simulation time, $T$.

Have $N - m$ runs where the particle leaves at time $t_i$.

Pretend that in each leaving run, I stopped watching when the particle left. This means the total observation time is,

$T_\mathrm{tot} = m T + \sum_i t_i$

So have $N - m$ hops in time $T_\mathrm{tot}$,

$\bar{\tau} = \dfrac{T_\mathrm{tot}}{N - m} = \dfrac{m T + \sum_i t_i}{N - m}$
